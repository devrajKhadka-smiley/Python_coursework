```python
mat_df = pd.read_csv("datascience.csv")
mat_df
```

## having no data consistency in data| making a consistent data


```python
mat_df['job_title'].replace('Data Analytics Manager','Data Analyst', inplace = True)
mat_df['job_title'].replace('ML Engineer','Machine Learning Engineer', inplace = True)
mat_df['job_title'].replace('Data Science Manager','Data Science', inplace = True)
mat_df['job_title'].replace('Data Science Consultant','Data Science', inplace = True)


# data analytics manager = data analytics
# ml engineer = machine learnig Engineer
# data science manager = data science Consultant = data science
```

```python
top_15_jobs = mat_df['job_title'].value_counts().nlargest(15)
top_15_jobs
```

```python
# de = pd.read_csv("datascience.csv")
# top_15_jobs = de['job_title'].value_counts().nlargest(15)

colors = ['red', 'blue', 'green', 'yellow', 'magenta', 'maroon', 'lime', 'cyan', 'olive', 'purple', 'black', 'gray', 'pink', 'brown', 'violet']

plt.figure(figsize=(10,9))
bars = plt.bar(top_15_jobs.index, top_15_jobs, color=colors)

plt.bar_label(bars, fmt='%d')

plt.xlabel('Jobs')
plt.ylabel('Count')
plt.title('Top 15 Jobs')

plt.xticks([])

legend_labels = dict(zip(top_15_jobs.index, colors))
legend_handles = [plt.Rectangle((0,0),1,1, color=color) for color in colors]
plt.legend(legend_handles, legend_labels.keys(), loc='upper right')

plt.show()
```

## Write a python program to find out salaries based on experience level.Illustrate it through bar graph.

```python
exp_pay = mat_df.groupby('experience_level')['salary'].mean()
exp_pay
```

```python
dd = pd.read_csv("datascience.csv")
exp_salary  = dd.groupby('experience_level')['salary'].mean()

color = ['red', 'blue', 'green', 'yellow']

plt.figure(figsize=(5,9))
bars = plt.bar(exp_salary.index, exp_salary, color = colors)

plt.bar_label(bars, fmt='%d')

plt.xlabel('Jobs')
plt.ylabel('Count')
plt.title("Salary Based on Experience")

plt.xticks([])

legend_labels = dict(zip(exp_salary.index, colors))
legend_handles = [plt.Rectangle((0,0),1,1, color=color) for color in colors]
plt.legend(legend_handles, legend_labels.keys(), loc='upper right')
plt.show()
```
