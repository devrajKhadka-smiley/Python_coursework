## well no data consistency is found | making a consistent data

```python
df['job_title'].replace('Data Analytics Manager','Data Analyst', inplace = True)
df['job_title'].replace('ML Engineer','Machine Learning Engineer', inplace = True)
df['job_title'].replace('Data Science Manager','Data Science', inplace = True)
df['job_title'].replace('Data Science Consultant','Data Science', inplace = True)


# data analytics manager = data analytics
# ml engineer = machine learnig Engineer
# data science manager = data science Consultant = data science
```

```python
top_15_jobs = df['job_title'].value_counts().nlargest(15)
top_15_jobs
```

```python
colors = ['red', 'blue', 'green', 'yellow', 'magenta', 'maroon', 'lime', 'cyan', 'olive', 'purple', 'black', 'gray', 'pink', 'brown', 'violet']

plt.figure(figsize=(10,9))
bars = plt.bar(top_15_jobs.index, top_15_jobs, color=colors)

plt.bar_label(bars, fmt='%d')

plt.xlabel('Jobs')
plt.ylabel('Count')
plt.title('Top 15 Jobs')

plt.xticks([])

legend_labels = dict(zip(top_15_jobs.index, colors))
legend_handles = [plt.Rectangle((0,0),1,1, color=color) for color in colors]
plt.legend(legend_handles, legend_labels.keys(), loc='upper right')

plt.show()
```

## Which job has the highest salaries? Illustrate with bar graph.

```python
Highest_salary_in_jobs = df.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False).head(10)

colors = ['red', 'blue', 'green', 'yellow', 'magenta', 'maroon', 'lime', 'cyan', 'olive', 'purple', 'black', 'gray', 'pink', 'brown', 'violet']
plt.figure(figsize=(15,20))
bars = plt.bar(Highest_salary_in_jobs.index, Highest_salary_in_jobs.values, color=colors)

plt.bar_label(bars, fmt='%d')

plt.xlabel('Job List')
plt.ylabel('Highest Salary According to Job')
plt.title('Jobs With highest Salary')
plt.xticks([])

legend_labels = dict(zip(Highest_salary_in_jobs.index, colors))
legend_handles = [plt.Rectangle((0,0),1,1, color = color) for color in colors]
plt.legend(legend_handles, legend_labels.keys(), loc='upper right')


plt.show()
```

## Write a python program to find out salaries based on experience level.Illustrate it through bar graph.

```python
exp_pay = df.groupby('experience_level')['salary_in_usd'].mean().sort_values(ascending=True)
exp_pay
```

```python
color = ['red', 'blue', 'green', 'yellow']

plt.figure(figsize=(5,9))
bars = plt.bar(exp_pay.index, exp_pay, color = color)

plt.bar_label(bars, fmt='%d')

plt.xlabel('Jobs')
plt.ylabel('Count')
plt.title("Salary Based on Experience")

plt.xticks([])

legend_labels = dict(zip(exp_pay.index, color))
legend_handles = [plt.Rectangle((0,0),1,1, color=color) for color in color]
plt.legend(legend_handles, legend_labels.keys(), loc='upper left')
plt.show()
```

## Write a Python program to show histogram and box plot of any chosen different variables. Use proper labels in the graph.

```python

variable1 = 'salary_in_usd'
variable2 = 'experience_level'

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x=variable1, kde=True, color='skyblue')
plt.title('Histogram of ' + variable1)
plt.xlabel(variable1)
plt.ylabel('Frequency')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x=variable2, y=variable1, palette='Set2')
plt.title('Box Plot of ' + variable1 + ' by ' + variable2)
plt.xlabel(variable2)
plt.ylabel(variable1)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data = [df[df[variable2] == category][variable1] for category in df[variable2].unique()]
plt.boxplot(data)
plt.title('Box Plot of ' + variable1 + ' by ' + variable2)
plt.xlabel(variable2)
plt.ylabel(variable1)
plt.show()
```
